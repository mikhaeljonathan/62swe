# Learning Rate

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data points roughly fitted into a line
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 100, 50)
y = 2.5 * x + np.random.normal(0, 20, size=x.shape)

# Create scatterplot
plt.scatter(x, y)
plt.title('Scatterplot of Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Learning rate too high

```python
# Initialize weights and parameters
weight = 0.0  # Initial weight (slope)
bias = 0.0    # Initial bias (intercept)
learning_rate = 0.001  # Small learning rate for gradient descent
iterations = 10  # Number of iterations

# Function to calculate mean squared error (MSE)
def calculate_mse(x, y, weight, bias):
    predictions = weight * x + bias
    mse = np.mean((y - predictions) ** 2)
    return mse

# Function to perform gradient descent
def gradient_descent(x, y, weight, bias, learning_rate):
    n = len(y)
    predictions = weight * x + bias
    weight_gradient = (-1/n) * np.sum(x * (y - predictions))
    bias_gradient = (-1/n) * np.sum(y - predictions)
    weight -= learning_rate * weight_gradient
    bias -= learning_rate * bias_gradient
    return weight, bias

# Perform gradient descent and track MSE for each epoch
mse_history = []
for iteration in range(iterations):
    weight, bias = gradient_descent(x, y, weight, bias, learning_rate)
    mse = calculate_mse(x, y, weight, bias)
    mse_history.append(mse)
    if iteration % 1 == 0:
      print(f'Iteration {iteration + 1}: MSE = {mse}')
      plt.plot(x, weight * x + bias, label=f'Iteration {iteration}')
      plt.scatter(x, y, alpha=0.5)
      plt.show()

# Plot MSE history over epochs
plt.plot(mse_history)
plt.title('MSE over Iteration')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.show()

# Final weight and bias after gradient descent
weight, bias
```

# Train test split

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Generate random data points roughly fitted into a line
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 100, 50)
y = 2.5 * x + np.random.normal(0, 20, size=x.shape)

# Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize weights and parameters for training
weight = 0.0  # Initial weight (slope)
bias = 0.0    # Initial bias (intercept)
learning_rate = 0.0001  # Small learning rate for gradient descent
iterations = 100  # Number of iterations

# Function to calculate mean squared error (MSE) for training and testing
def calculate_mse(x, y, weight, bias):
    predictions = weight * x + bias
    mse = np.mean((y - predictions) ** 2)
    return mse

# Function to perform gradient descent
def gradient_descent(x, y, weight, bias, learning_rate):
    n = len(y)
    predictions = weight * x + bias
    weight_gradient = (1/n) * np.sum(x * (predictions - y))
    bias_gradient = (1/n) * np.sum(predictions - y)
    weight -= learning_rate * weight_gradient
    bias -= learning_rate * bias_gradient
    return weight, bias

# Perform gradient descent and track MSE for both training and test sets
train_mse_history = []
test_mse_history = []
for iteration in range(iterations):
    weight, bias = gradient_descent(x_train, y_train, weight, bias, learning_rate)

    # Calculate MSE for training and test sets
    train_mse = calculate_mse(x_train, y_train, weight, bias)
    test_mse = calculate_mse(x_test, y_test, weight, bias)

    # Store MSE for each iteration
    train_mse_history.append(train_mse)
    test_mse_history.append(test_mse)

    # Optional: Display plots every 10,000 iterations
    if iteration % 10 == 0:
        print(f'Iteration {iteration + 1}: Train MSE = {train_mse}, Test MSE = {test_mse}')
        plt.plot(x_train, weight * x_train + bias, label=f'Iteration {iteration}')
        plt.scatter(x_train, y_train, alpha=0.5, label="Train")
        plt.scatter(x_test, y_test, alpha=0.5, color='red', label="Test")
        plt.legend()
        plt.show()

# Plot MSE history for both training and test sets
plt.plot(train_mse_history, label='Train MSE')
plt.plot(test_mse_history, label='Test MSE')
plt.title('MSE over Iterations (Train vs Test)')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.show()

# Final weight and bias after gradient descent
weight, bias
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# Generate more complex data (cubic relationship with some noise)
np.random.seed(42)
x_complex = np.linspace(0, 100, 50)
y_complex = 0.01 * x_complex**3 - 0.5 * x_complex**2 + 2.5 * x_complex + np.random.normal(0, 1000, size=x_complex.shape)

# Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x_complex, y_complex, test_size=0.2, random_state=42)

# Prepare polynomial features (degree 3)
poly = PolynomialFeatures(degree=3)
x_train_poly = poly.fit_transform(x_train.reshape(-1, 1))
x_test_poly = poly.fit_transform(x_test.reshape(-1, 1))

# Initialize weights and parameters for training
weights = np.zeros(x_train_poly.shape[1])  # Initialize all weights for polynomial features
learning_rate = 0.0000000001  # Small learning rate for gradient descent
iterations = 1000  # Number of iterations

# Function to calculate mean squared error (MSE) for polynomial regression
def calculate_mse_poly(x_poly, y, weights):
    predictions = x_poly @ weights  # Matrix multiplication for polynomial regression
    mse = np.mean((y - predictions) ** 2)
    return mse

# Function to perform gradient descent for polynomial regression
def gradient_descent_poly(x_poly, y, weights, learning_rate):
    n = len(y)
    predictions = x_poly @ weights  # Matrix multiplication for predictions
    weights_gradient = (-1/n) * (x_poly.T @ (y - predictions))  # Gradient calculation
    weights -= learning_rate * weights_gradient
    return weights

# Perform gradient descent and track MSE for both training and test sets
train_mse_history = []
test_mse_history = []
for iteration in range(iterations):
    weights = gradient_descent_poly(x_train_poly, y_train, weights, learning_rate)

    # Calculate MSE for training and test sets
    train_mse = calculate_mse_poly(x_train_poly, y_train, weights)
    test_mse = calculate_mse_poly(x_test_poly, y_test, weights)

    # Store MSE for each iteration
    train_mse_history.append(train_mse)
    test_mse_history.append(test_mse)

    # Optional: Display plots every 10,000 iterations
    if iteration % 100 == 0:
        print(f'Iteration {iteration + 1}: Train MSE = {train_mse}, Test MSE = {test_mse}')
        plt.scatter(x_train, y_train, alpha=0.5, label="Train")
        plt.scatter(x_test, y_test, alpha=0.5, color='red', label="Test")
        plt.plot(np.sort(x_train), np.sort(x_train_poly @ weights), label=f'Iteration {iteration}', color='green')
        plt.legend()
        plt.show()

# Plot MSE history for both training and test sets
plt.plot(train_mse_history, label='Train MSE')
plt.plot(test_mse_history, label='Test MSE')
plt.title('MSE over Iterations (Train vs Test)')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.show()

# Final weights after gradient descent
weights
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error

# Generate more complex data (cubic relationship with some noise)
np.random.seed(42)
x_complex = np.linspace(0, 100, 50)
y_complex = 0.01 * x_complex**3 - 0.5 * x_complex**2 + 2.5 * x_complex + np.random.normal(0, 1000, size=x_complex.shape)

# Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x_complex, y_complex, test_size=0.2, random_state=42)

# Prepare polynomial features (degree 3)
poly = PolynomialFeatures(degree=3)
x_train_poly = poly.fit_transform(x_train.reshape(-1, 1))
x_test_poly = poly.fit_transform(x_test.reshape(-1, 1))

# Initialize weights and parameters for training
initial_weights = np.zeros(x_train_poly.shape[1])

# Define the loss function (MSE) for optimization
def loss_function(weights, x_poly, y):
    predictions = x_poly @ weights
    mse = np.mean((y - predictions) ** 2)
    return mse

# Use Scipy's minimize function to perform gradient descent and optimize weights
result = minimize(loss_function, initial_weights, args=(x_train_poly, y_train), method='BFGS')

# Get the optimized weights
optimized_weights = result.x

# Predict on the training and test sets using the optimized weights
y_train_pred = x_train_poly @ optimized_weights
y_test_pred = x_test_poly @ optimized_weights

# Calculate MSE for training and test sets
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

# Print the MSE results
print(f'Train MSE: {train_mse}')
print(f'Test MSE: {test_mse}')

# Plot the original data and the polynomial regression line
plt.scatter(x_train, y_train, alpha=0.5, label="Train")
plt.scatter(x_test, y_test, alpha=0.5, color='red', label="Test")
plt.plot(np.sort(x_train), np.sort(y_train_pred), label='Polynomial Regression (Train)', color='green')
plt.legend()
plt.show()

# Final optimized weights after optimization
print(f'Optimized weights: {optimized_weights}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error

# Generate more complex data (cubic relationship with some noise)
np.random.seed(42)
x_complex = np.linspace(0, 100, 50)
y_complex = 0.01 * x_complex**3 - 0.5 * x_complex**2 + 2.5 * x_complex + np.random.normal(0, 1000, size=x_complex.shape)

# Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x_complex, y_complex, test_size=0.2, random_state=42)

# Prepare polynomial features (degree 3)
poly = PolynomialFeatures(degree=3)
x_train_poly = poly.fit_transform(x_train.reshape(-1, 1))
x_test_poly = poly.fit_transform(x_test.reshape(-1, 1))

# Initialize weights and parameters for training
initial_weights = np.zeros(x_train_poly.shape[1])

# Define the regularization parameter
alpha = 1  # Adjust this value to control the amount of regularization

# Define the loss function (MSE + L2 Regularization) for optimization
def loss_function(weights, x_poly, y, alpha):
    predictions = x_poly @ weights
    mse = np.mean((y - predictions) ** 2)
    l2_regularization = alpha * np.sum(weights ** 2)
    return mse + l2_regularization

# Use Scipy's minimize function to perform gradient descent and optimize weights with regularization
result = minimize(loss_function, initial_weights, args=(x_train_poly, y_train, alpha), method='BFGS')

# Get the optimized weights
optimized_weights = result.x

# Predict on the training and test sets using the optimized weights
y_train_pred = x_train_poly @ optimized_weights
y_test_pred = x_test_poly @ optimized_weights

# Calculate MSE for training and test sets
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

# Print the MSE results
print(f'Train MSE: {train_mse}')
print(f'Test MSE: {test_mse}')

# Plot the original data and the polynomial regression line
plt.scatter(x_train, y_train, alpha=0.5, label="Train")
plt.scatter(x_test, y_test, alpha=0.5, color='red', label="Test")
plt.plot(np.sort(x_train), np.sort(y_train_pred), label='Polynomial Regression (Train)', color='green')
plt.legend()
plt.show()

# Final optimized weights after optimization
print(f'Optimized weights with regularization: {optimized_weights}')
```

# Iteration using tensorflow

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# Generate more complex data (cubic relationship with some noise)
np.random.seed(42)
x_complex = np.linspace(0, 100, 50)
y_complex = 0.01 * x_complex**3 - 0.5 * x_complex**2 + 2.5 * x_complex + np.random.normal(0, 1000, size=x_complex.shape)

# Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x_complex, y_complex, test_size=0.2, random_state=42)

# Normalize the data using StandardScaler (works similar to your previous normalization)
scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train.reshape(-1, 1))
x_test_norm = scaler.transform(x_test.reshape(-1, 1))

# Build the neural network
model = Sequential()
model.add(Dense(64, input_dim=1, activation='relu'))  # Hidden layer with 64 units and ReLU activation
model.add(Dense(64, activation='relu'))               # Another hidden layer
model.add(Dense(1, activation='linear'))              # Output layer with linear activation for regression

# Compile the model with mean squared error as the loss function and Adam optimizer
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.001))

# Track training and test MSE at each epoch
train_mse_history = []
test_mse_history = []

# Training the model iteratively to capture MSE over epochs
epochs = 1000
batch_size = 10
for epoch in range(epochs):
    # Train the model for one epoch
    model.fit(x_train_norm, y_train, batch_size=batch_size, verbose=0)

    # Calculate the MSE for training and test sets
    train_mse = model.evaluate(x_train_norm, y_train, verbose=0)
    test_mse = model.evaluate(x_test_norm, y_test, verbose=0)

    # Store MSE history
    train_mse_history.append(train_mse)
    test_mse_history.append(test_mse)

    # Optional: Display plots every 100 epochs
    if epoch % 100 == 0:
        print(f'Epoch {epoch + 1}: Train MSE = {train_mse}, Test MSE = {test_mse}')
        plt.scatter(x_train_norm, y_train, alpha=0.5, label="Train")
        plt.scatter(x_test_norm, y_test, alpha=0.5, color='red', label="Test")
        plt.plot(np.sort(x_train_norm, axis=0), np.sort(model.predict(x_train_norm), axis=0), label=f'Epoch {epoch}', color='green')
        plt.legend()
        plt.show()

# Plot MSE history for both training and test sets
plt.plot(train_mse_history, label='Train MSE')
plt.plot(test_mse_history, label='Test MSE')
plt.title('MSE over Epochs (Train vs Test)')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.show()

# Final weights and biases
weights, biases = model.layers[0].get_weights()  # Fetching the weights of the first layer
weights, biases
```

# Cross validation with gridsearch

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, KFold
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error

# Generate more complex data (cubic relationship with some noise)
np.random.seed(42)
x_complex = np.linspace(0, 100, 50)
y_complex = 0.01 * x_complex**3 - 0.5 * x_complex**2 + 2.5 * x_complex + np.random.normal(0, 1000, size=x_complex.shape)

# Split the data into training and test sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x_complex, y_complex, test_size=0.2, random_state=42)

# Prepare polynomial features (degree 3)
poly = PolynomialFeatures(degree=3)
x_train_poly = poly.fit_transform(x_train.reshape(-1, 1))
x_test_poly = poly.fit_transform(x_test.reshape(-1, 1))

# Define the loss function (MSE + L2 Regularization) for optimization
def loss_function(weights, x_poly, y, alpha):
    predictions = x_poly @ weights
    mse = np.mean((y - predictions) ** 2)
    l2_regularization = alpha * np.sum(weights ** 2)
    return mse + l2_regularization

# Grid search over a range of alpha values with cross-validation
alphas = np.logspace(-4, 1, 10)  # Adjust the range of alpha values as needed
kf = KFold(n_splits=5, shuffle=True, random_state=42)
best_alpha = None
best_cv_mse = float("inf")

for alpha in alphas:
    cv_mse = []
    for train_index, val_index in kf.split(x_train_poly):
        x_cv_train, x_cv_val = x_train_poly[train_index], x_train_poly[val_index]
        y_cv_train, y_cv_val = y_train[train_index], y_train[val_index]

        # Initialize weights for each fold
        initial_weights = np.zeros(x_cv_train.shape[1])

        # Optimize weights using the minimize function
        result = minimize(loss_function, initial_weights, args=(x_cv_train, y_cv_train, alpha), method='BFGS')
        optimized_weights = result.x

        # Predict on the validation set and calculate MSE
        y_cv_val_pred = x_cv_val @ optimized_weights
        fold_mse = mean_squared_error(y_cv_val, y_cv_val_pred)
        cv_mse.append(fold_mse)

    # Average MSE across folds for the current alpha
    avg_cv_mse = np.mean(cv_mse)
    print(f'Alpha: {alpha}, CV MSE: {avg_cv_mse}')

    # Update the best alpha if current alpha has a lower average MSE
    if avg_cv_mse < best_cv_mse:
        best_cv_mse = avg_cv_mse
        best_alpha = alpha

print(f'\nBest alpha found: {best_alpha} with CV MSE: {best_cv_mse}')

# Retrain on full training set with the best alpha
initial_weights = np.zeros(x_train_poly.shape[1])
result = minimize(loss_function, initial_weights, args=(x_train_poly, y_train, best_alpha), method='BFGS')
optimized_weights = result.x

# Predict on the training and test sets using the optimized weights
y_train_pred = x_train_poly @ optimized_weights
y_test_pred = x_test_poly @ optimized_weights

# Calculate MSE for training and test sets
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

# Print the MSE results
print(f'Train MSE: {train_mse}')
print(f'Test MSE: {test_mse}')

# Plot the original data and the polynomial regression line
plt.scatter(x_train, y_train, alpha=0.5, label="Train")
plt.scatter(x_test, y_test, alpha=0.5, color='red', label="Test")
plt.plot(np.sort(x_train), np.sort(y_train_pred), label='Polynomial Regression (Train)', color='green')
plt.legend()
plt.show()

# Final optimized weights after optimization
print(f'Optimized weights with best alpha ({best_alpha}): {optimized_weights}')
```
