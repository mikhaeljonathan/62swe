# Basic plotting

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Simple Line Plot')

# Display the plot
plt.show()
```

# Plot types

* Line Plot: plt.plot()
* Scatter Plot: plt.scatter()
* Bar Plot: plt.bar()
* Histogram: plt.hist()
* Pie Chart: plt.pie()

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 1, 2, 3, 4, 5]
y = [5, 7, 6, 2, 3, 13]

# Create scatter plot
plt.scatter(x, y)

# Display plot
plt.show()
```

```python
plt.bar(x, y)
```

```python
plt.hist(x)
```

```python
import matplotlib.pyplot as plt

# Sample data representing parts of a whole
x_pie = [20, 35, 15, 30]  # Example values for each category
labels = ['Category A', 'Category B', 'Category C', 'Category D']  # Names for each slice
colors = ['skyblue', 'lightcoral', 'yellowgreen', 'gold']  # Custom colors for each slice
explode = [0, 0.1, 0, 0]  # Emphasize the second slice (Category B)

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    x_pie,
    labels=labels,           # Add labels for each slice
    autopct='%1.1f%%',       # Show percentages with one decimal place
    startangle=90,           # Start the first slice at the top (90 degrees)
    colors=colors,           # Use custom colors
    explode=explode,         # "Explode" one of the slices for emphasis
    wedgeprops={'edgecolor': 'black'}  # Add border to each slice for clarity
)

# Add title
plt.title("Distribution of Categories in Data")

# Display the chart
plt.show()
```

# Subplots

```python
fig, axs = plt.subplots(1, 2)  # 1 row, 2 columns

# First subplot
axs[0].plot(x, y)
axs[0].set_title('Plot 1')

# Second subplot
axs[1].scatter(x, y)
axs[1].set_title('Plot 2')

fig.savefig("plot.png")

plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y)
axs[1, 0].bar(x, y)
axs[1, 1].hist(y)

plt.show()
```

Seaborn is essentially a higher-level API for Matplotlib. It simplifies the process of creating complex visualizations by offering a more intuitive syntax, and it takes care of common data visualization tasks automatically (like improving the aesthetics of the plot).

Seaborn and Matplotlib share the same state by default, meaning that when you use plt.figure() to set the size of the figure or add a title with plt.title(), it will affect the plot generated by Seaborn because they are modifying the same underlying Matplotlib figure and axes.

```python
import seaborn as sns
import pandas as pd

# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Title 1")

data = {'x': [1, 2, 3, 4, 5], 'y': [2, 3, 5, 7, 11]}
df = pd.DataFrame(data)

# sns.lineplot(data=data)
sns.lineplot(x='x', y='y', data=data)
plt.show()
```

```python
import numpy as np

data1 = np.random.normal(loc=0, scale=1, size=1000)   # Normal distribution centered at 0
data2 = np.random.normal(loc=3, scale=1.5, size=1000) # Normal distribution centered at 3, wider spread
```

```python
data1
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data1, bins=30, kde=True, color='skyblue')
plt.title("Histogram with KDE Overlay")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data1, bins=40, color='lightgreen', stat='density')
sns.kdeplot(data1, color='darkgreen', linewidth=2, fill=True)
plt.title("Customized Histogram with KDE")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data1, color='blue', fill=True, label='Data 1')
sns.kdeplot(data2, color='red', fill=True, label='Data 2')
plt.title("KDE Plot of Two Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
