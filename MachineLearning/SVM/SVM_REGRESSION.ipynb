# Support Vector Machine - Support Vector Regressor

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.datasets import make_regression

# Step 1: Generate a non-linear dataset (for regression)
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)  # 80 data points between 0 and 5
y = np.sin(X).ravel() + 0.1 * np.random.randn(80)  # Adding some noise to the sine function

# Step 2: Create the Support Vector Regressor model (using RBF kernel by default)
svr_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_model.fit(X, y)

# Step 3: Predict using the trained SVR model
X_test = np.linspace(0, 5, 100).reshape(-1, 1)  # Test points for prediction
y_pred = svr_model.predict(X_test)

# Step 4: Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='red', label='Data points')
plt.plot(X_test, y_pred, color='blue', label='SVR fit')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Support Vector Regression (SVR) with RBF Kernel')
plt.legend()
plt.show()
```
