# Support Vector Machine - RBF Kernel

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_circles

# Step 1: Generate non-linearly separable data (two classes in a circular pattern)
X, y = make_circles(n_samples=100, factor=0.5, noise=0.1)

# Step 2: Create and train the SVM model with a Gaussian (RBF) kernel
# C parameter controls the trade-off between achieving a low error on the training data and a low-complexity model
svm_model = SVC(kernel='rbf', gamma=1, C=1)  # gamma is related to the Gaussian kernel width (sigma)
svm_model.fit(X, y)

# Step 3: Visualize the decision boundary
# Create a meshgrid for plotting decision regions
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Predict the label for each point in the meshgrid
Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the results
plt.contourf(xx, yy, Z, alpha=0.75, cmap=plt.cm.coolwarm)  # Background color for decision regions
plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', edgecolors='k', cmap=plt.cm.coolwarm)  # Scatter plot of data points
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM with Gaussian Kernel (Non-Linearly Separable Data)')
plt.show()
```
